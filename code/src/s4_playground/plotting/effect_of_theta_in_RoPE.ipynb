```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ggplot")
d = 1024
base = 10_000
theta = base**(-2*(np.arange(1,(d+1)/2) -1)/d)

plt.plot(range(1, d, 2), theta, label="base: 10000")
base = 10
theta = base**(-2*(np.arange(1,(d+1)/2) -1)/d)
plt.plot(range(1, d, 2), theta, label="base: 10")
plt.ylabel(r"$\theta_i$", rotation=0)

plt.xlabel("d: model dimension")
plt.title(r"$\Theta=\left\{\theta_i=base^{-2(i-1) / d}, i \in[1,2, \ldots, d / 2]\right\}$")
plt.legend()
plt.tight_layout()
plt.savefig("data/exp2/RoPE_theta.png", dpi=200)
plt.show()
print(len(theta))


```
